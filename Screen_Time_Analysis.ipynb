```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
df1 = pd.read_csv("Screentime-App-Details.csv")
df1.head() #Displays the first 5 rows of the dataframe
```

```python
pd.pivot_table(df1, index= 'App', values= "Times opened")
```

```python
pd.pivot_table(df1, index= 'Date',columns= 'App', values= "Usage")
```

```python
pd.pivot_table(df1, index= 'Date',columns= 'App', values= "Notifications")
```

```python
pd.pivot_table(df1, index= 'Date',columns= 'App', values= "Times opened")
```

```python
pd.pivot_table(df1, index= 'Date', columns= 'App', values= "Usage").plot(
    kind= 'bar',
    figsize=(15,5),
    color=['pink','green'],
    title="Usage",
    xlabel='Date',ylabel='Usage')
```

```python
pd.pivot_table(df1, index= 'Date', columns= 'App', values= "Notifications").plot(
    kind= 'bar',
    figsize=(15,5),
    color=['pink','green'],
    title="Notifications",
    xlabel='Date',ylabel='Number of notifications')
```

```python
pd.pivot_table(df1, index= 'Date', columns= 'App', values= "Times opened").plot(
    kind= 'bar',
    figsize=(15,5),
    color=['pink','green'],
    title="Times opened",
    xlabel='Date',ylabel='Times opened')
```

```python
figure = px.bar(data_frame=df1,
                x = "Date",
                y = "Usage",
                color="App",
                color_discrete_sequence=[
                 "pink", "green"],
                title="Usage")
figure.show()
```

```python
figure = px.bar(data_frame=df1,
                x = "Date",
                y = "Notifications",
                color="App",
                color_discrete_sequence=[
                 "pink", "green"],
                title="Notifications")
figure.show()
```

```python
figure = px.bar(data_frame=df1,
                x = "Date",
                y = "Times opened",
                color="App",
                color_discrete_sequence=[
                 "pink", "green"],
                title="Times Opened")
figure.show()
```

```python
import seaborn as sb
dataplot=sb.heatmap(df1.corr(),
                    cmap='Blues')

# displaying heatmap
plt.show()
```

```python
figure = px.scatter(data_frame = df1,
                    x="Notifications",
                    y="Usage",
                    size="Notifications",
                    trendline="ols",
                    color_discrete_sequence=["purple"],
                    title = "Relationship Between Number of Notifications and Usage")
figure.show()
```

```python
figure = px.scatter(data_frame = df1,
                    x="Notifications",
                    y="Times opened",
                    size="Notifications",
                    trendline="ols",
                    color_discrete_sequence=["purple"],
                    title = "Relationship Between Number of Notifications and Times Opened")
figure.show()
```
